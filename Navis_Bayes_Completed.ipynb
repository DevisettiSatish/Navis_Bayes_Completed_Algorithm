```python
# since all are categorical features we make them into letters:
```

```python
# outLook > rainy=0,overcast=1,sunny=2
```

```python

a['OutLook'] = a['OutLook'].map({'rainy':0 , 'overcast':1 , 'sunny':2})
a
```

```python
a['Temperature']=a['Temperature'].map({'hot':0,'mild':1,'cool':2})
a
```

```python
a['Humidity']=a['Humidity'].map({'high':1,'normal':2})
a
```

```python
a['windy']=a['windy'].map({'True':0,'False':2})
a
```

```python
a['sports']=a['sports'].map({'no':0,'yes':1})
a
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
reg=GaussianNB()
```

```python
X = a.iloc[: , :-1]
y = a.iloc[: , -1]
```

```python
reg.fit(X,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.22,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
y_train_pred=reg.predict(X_train)
y_train_pred
```

```python
training_data=pd.DataFrame(X_train)
training_data
```

```python
training_data['y_train_model_prediction'] = y_train_pred
training_data
```

```python
test_data=pd.DataFrame(X_test)
test_data
```

```python
y_test_pred=reg.predict(X_test)
y_test_pred
```

```python
test_data['y_test_model_prediction']=y_test_pred
test_data
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# train_Data
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
accuracy_score(y_train,y_train_pred)
```

```python
print(classification_report(y_train,y_train_pred))
```

```python
# test_Data
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
accuracy_score(y_test,y_test_pred)
```
